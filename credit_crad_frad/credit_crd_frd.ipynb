```python
dt = pd.to_datetime(df["unix_time"], unit="s")

df["hour"] = dt.dt.hour
df["dayofweek"] = dt.dt.dayofweek
df["is_weekend"] = dt.dt.dayofweek.isin([5, 6]).astype(int)
```

```python
import numpy as np

# Earth radius in km
R = 6371

lat1 = np.radians(df["lat"])
lon1 = np.radians(df["long"])
lat2 = np.radians(df["merch_lat"])
lon2 = np.radians(df["merch_long"])

dlat = lat2 - lat1
dlon = lon2 - lon1

a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2

df["cust_merch_dist"] = 2 * R * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
cat_cols = ["category", "gender", "state"]
num_cols = [col for col in X_train.columns if col not in cat_cols]
```

```python
X.head()
```

```python
y.head()
```

```python
num_cols
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(drop="first", handle_unknown="ignore"), cat_cols),
        ("num", "passthrough", num_cols)
    ]
)
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

```python
model = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("clf", LogisticRegression(
        max_iter=1000,
        class_weight="balanced",
        n_jobs=-1
    ))
])
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)

y_proba = model.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(cm).plot()
plt.show()
```

```python
from sklearn.metrics import roc_auc_score

roc_auc = roc_auc_score(y_test, y_proba)
print("ROC-AUC:", roc_auc)
```

```python
custom_pred = (y_proba > 0.3).astype(int)
print(classification_report(y_test, custom_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
rf_model = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("clf", RandomForestClassifier(
        n_estimators=200,
        max_depth=None,
        min_samples_split=5,
        class_weight="balanced",
        random_state=42,
        n_jobs=-1
    ))
])
```

Can we improve our Random Forest by tuning its hyperparameters using cross-validation?

❓ Why NOT GridSearchCV?

Dataset size: 1M+ rows

GridSearch = tries every combination ❌ (too slow)

RandomizedSearchCV = smart sampling ✅ (industry choice)

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_dist = {
    "clf__n_estimators": [100, 200, 300],
    "clf__max_depth": [None, 10, 20, 30],
    "clf__min_samples_split": [2, 5, 10],
    "clf__min_samples_leaf": [1, 2, 4],
    "clf__max_features": ["sqrt", "log2"]
}
```

Hyperparameter tuning using RandomizedSearchCV improved the Random Forest’s ROC-AUC and fraud recall, making it more effective for real-world fraud detection

```python
rf_search = RandomizedSearchCV(
    estimator=rf_model,
    param_distributions=param_dist,
    n_iter=10,
    scoring="roc_auc",
    cv=3,
    verbose=2,
    n_jobs=-1,
    random_state=42
)
```

```python
rf_search.fit(X_train, y_train)
```

Hyperparameter tuning using RandomizedSearchCV was attempted. Due to the large dataset size and high computational cost, full tuning was stopped after observing long execution time. The baseline Random Forest already achieved excellent performance, so it was retained for further comparison.

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib


model = joblib.load("fraud_xgb_model.pkl")

st.set_page_config(page_title="Credit Card Fraud Detection", layout="centered")

st.title("💳 Credit Card Fraud Detection")
st.write("Enter transaction details to predict fraud probability")


category = st.selectbox("Category", [
    "gas_transport", "grocery_pos", "shopping_pos", "shopping_net",
    "entertainment", "food_dining", "misc_pos", "misc_net"
])

gender = st.selectbox("Gender", ["M", "F"])
state = st.text_input("State (e.g. CA, NY)", "CA")

zip_code = st.number_input("ZIP Code", min_value=0, value=12345)
lat = st.number_input("Customer Latitude", value=40.0)
long = st.number_input("Customer Longitude", value=-75.0)
city_pop = st.number_input("City Population", min_value=0, value=100000)

merch_lat = st.number_input("Merchant Latitude", value=41.0)
merch_long = st.number_input("Merchant Longitude", value=-74.0)

log_amt = st.number_input("Log Amount", value=3.5)
hour = st.slider("Transaction Hour", 0, 23, 12)
dayofweek = st.slider("Day of Week (0=Mon)", 0, 6, 3)
is_weekend = st.selectbox("Is Weekend?", [0, 1])
cust_merch_dist = st.number_input("Customer–Merchant Distance (km)", value=50.0)

input_df = pd.DataFrame([{
    "category": category,
    "gender": gender,
    "state": state,
    "zip": zip_code,
    "lat": lat,
    "long": long,
    "city_pop": city_pop,
    "merch_lat": merch_lat,
    "merch_long": merch_long,
    "log_amt": log_amt,
    "hour": hour,
    "dayofweek": dayofweek,
    "is_weekend": is_weekend,
    "cust_merch_dist": cust_merch_dist
}])

if st.button("Predict Fraud"):
    proba = model.predict_proba(input_df)[0][1]
    prediction = "🚨 FRAUD" if proba > 0.3 else "✅ Not Fraud"

    st.subheader("Result")
    st.write(f"**Fraud Probability:** `{proba:.6f}`")
    st.write(f"**Prediction:** {prediction}")
```
